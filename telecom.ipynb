# split data


```python
from sklearn.model_selection import train_test_split
X = train.drop('churn_probability', axis=1)
y = train['churn_probability']
# train['churn_probability']
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
def bar_chart():
    train_class0 = pd.Series(y_train[y_train == 0]).value_counts()
    train_class1 = pd.Series(y_train[y_train == 1]).value_counts()

    test_class0 = pd.Series(y_test[y_test == 0]).value_counts()
    test_class1 = pd.Series(y_test[y_test == 1]).value_counts()

    df_train = pd.DataFrame({'Class 0': train_class0, 'Class 1': train_class1})
    df_test = pd.DataFrame({'Class 0': test_class0, 'Class 1': test_class1})

    # Plotting
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    df_train.plot(kind='bar', stacked=True, ax=ax1, title='Class Distribution in Train Set')
    df_test.plot(kind='bar', stacked=True, ax=ax2, title='Class Distribution in Test Set')

    ax1.set_ylabel('Count')
    ax2.set_ylabel('Count')

    fig.suptitle('Stacked Bar Chart for Class Distribution between Train and Test Sets')
bar_chart()
```
